```python
nbr_match = int(input("A combien de match souhaiter vous assister"))
tarif = int(input(" Choissisez entre le tarif 1, 2 ou 3 (taper 1, 2 ou 3)"))
total = 0
if tarif == 1:
    total =  8*nbr_match
elif tarif == 2:
    total = 44 + (4*nbr_match)
else:
    total = 91

print("vous devez payer un total de",total, "€")
```
